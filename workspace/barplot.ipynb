```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 数据参数
lenofX = 6 # 横坐标数量
dataX = { # 不用管
    'ax00':np.arange(lenofX),
    'ax01':np.arange(lenofX),
    'ax10':np.arange(lenofX),
    'ax11':np.arange(lenofX)
} 

dataY_Left = { #  修改为第一个柱子的值的序列
    'ax00':np.random.rand(len(dataX['ax00'])), 
    'ax01':np.random.rand(len(dataX['ax01'])),
    'ax10':np.random.rand(len(dataX['ax10'])),
    'ax11':np.random.rand(len(dataX['ax11'])),
}

dataY_Right = { # 修改为第二个柱子的值的序列
    'ax00':np.random.rand(len(dataX['ax00'])),
    'ax01':np.random.rand(len(dataX['ax01'])),
    'ax10':np.random.rand(len(dataX['ax10'])),
    'ax11':np.random.rand(len(dataX['ax11'])),
}

xTicks = { # 横轴的坐标
    'ax00': ['sheng\ncong', 'wo\ncao', 'niu\nbi','sheng\ncong', 'wo\ncao', 'niu\nbi666'],
    'ax01': ['sheng\ncong', 'wo\ncao', 'niu\nbi','sheng\ncong', 'wo\ncao', 'niu\nbi777'],
    'ax10': ['sheng\ncong', 'wo\ncao', 'niu\nbi','sheng\ncong', 'wo\ncao', 'niu\nbi888'],
    'ax11': ['sheng\ncong', 'wo\ncao', 'niu\nbi','sheng\ncong', 'wo\ncao', 'niu\nbi999']
}

xTicksFontDict = { # 横轴坐标的字体属性
    'fontsize': 7,
    'family': 'Times New Roman'
    
}

xLabels = { # 横轴名字
    'ax00':'Subplots aaaa',
    'ax01':'Subplots xx',
    'ax10':'Subplots sdad',
    'ax11':'Subplots good'
}
xLabelsFontDict = { # 横轴名字的字体属性
    'fontsize': 10,
    'family': 'Times New Roman'
}

yLabels = { # 纵轴名字
    'ax00':'Frequency',
    'ax01':'Frequency',
    'ax10':'Frequency',
    'ax11':'Frequency'
}
yLabelsFontDict = { # 纵轴名字的字体属性
    'fontsize': 10,
    'family': 'Times New Roman'
}

labelLegend0_L = 'Control' # 左边柱子的在图例中的名字
labelLegend0_R = 'Treatment' # 右边柱子的在图例中的名字
legendFontDict = { # 图例名字的字体属性
    'size': 10,
    'family': 'Times New Roman'
}

# 不管
axsNames = ['ax00', 'ax01', 'ax10', 'ax11']

yLims = { # 纵轴范围
    'ax00': (0, 1.5),
    'ax01': (0, 1.5),
    'ax10': (0, 1.5),
    'ax11': (0, 1.5),
}
yTicksDivided = 6 #纵轴分为几段

titles = {
    'ax00': 'Title good of ax00',
    'ax01': 'Title good of ax01',
    'ax10': 'Title good of ax10',
    'ax11': 'Title good of ax11',
}
titlesFontDict = { # 图例名字的字体属性
    'size': 12,
    'family': 'Times New Roman'
}
```

```python
# 绘图参数
barWidth = 0.3   # 柱子宽度
barOffset = barWidth/2.0 
barColor0 = '#EA3C17' # 柱子颜色 # 
barColor1 = '#54ACDD'
```

```python
fig, axs = plt.subplots(2, 2, figsize=(8,6), dpi=200) # 修改图比例和dpi
for idx in range(4):
    i = idx // 2
    j = idx % 2
    axName = axsNames[idx]
    axs[i][j].bar(dataX[axName]-barOffset, dataY_Left[axName],  width=barWidth, color=barColor0, label=labelLegend0_L)
    axs[i][j].bar(dataX[axName]+barOffset, dataY_Right[axName], width=barWidth, color=barColor1, label=labelLegend0_R)
    axs[i][j].set_xticks(dataX[axName])
    axs[i][j].set_xticklabels(xTicks[axName],fontdict=xTicksFontDict)
    axs[i][j].set_xlabel(xLabels[axName], fontdict=xLabelsFontDict)
    axs[i][j].set_ylabel(yLabels[axName], fontdict=yLabelsFontDict)
    axs[i][j].set_ylim(yLims[axName])
    YTICKS = np.linspace(yLims[axName][0], yLims[axName][1], yTicksDivided)
    axs[i][j].set_yticks(YTICKS)
    axs[i][j].set_yticklabels(['%.2f'%y for y in YTICKS],fontdict=yLabelsFontDict)
    axs[i][j].spines['right'].set_visible(False)
    axs[i][j].spines['top'].set_visible(False)
    axs[i][j].legend(frameon=False, prop=legendFontDict)
    axs[i][j].set_title(titles[axName], fontdict=titlesFontDict)


plt.subplots_adjust(hspace=0.5,wspace=0.25)  # 调整子图间距
```

```python
markers = ['o', 'o']
markerSize = 2
lineWidth = 1
lineStyles = ['--', '-']
```

```python
plt.style.use('default') # 可选 'ggplot'风格

fig, axs = plt.subplots(2, 2, figsize=(8,6), dpi=150)
for idx in range(4):
    i = idx // 2
    j = idx % 2
    axName = axsNames[idx]
    axs[i][j].plot(dataX[axName], dataY_Left[axName],  color=barColor0, label=labelLegend0_L,ls=lineStyles[0],lw=lineWidth, marker=markers[0], ms=markerSize)
    axs[i][j].plot(dataX[axName], dataY_Right[axName], color=barColor1, label=labelLegend0_R,ls=lineStyles[1],lw=lineWidth, marker=markers[0], ms=markerSize)
    axs[i][j].set_xticks(dataX[axName])
    axs[i][j].set_xticklabels(xTicks[axName],fontdict=xTicksFontDict)
    axs[i][j].set_xlabel(xLabels[axName], fontdict=xLabelsFontDict)
    axs[i][j].set_ylabel(yLabels[axName], fontdict=yLabelsFontDict)
    axs[i][j].set_ylim(yLims[axName])
    YTICKS = np.linspace(yLims[axName][0], yLims[axName][1], yTicksDivided)
    axs[i][j].set_yticks(YTICKS)
    axs[i][j].set_yticklabels(['%.2f'%y for y in YTICKS],fontdict=yLabelsFontDict)
    axs[i][j].spines['right'].set_visible(False)
    axs[i][j].spines['top'].set_visible(False)
    axs[i][j].legend(frameon=False, prop=legendFontDict)
    axs[i][j].set_title(titles[axName], fontdict=titlesFontDict)
    axs[i][j].grid(axis='y', lw=0.5, ls='--') 


plt.subplots_adjust(hspace=0.5,wspace=0.25)
```
